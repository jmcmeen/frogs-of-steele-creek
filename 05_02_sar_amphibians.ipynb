# The Species-Area Relationship of Amphibians in the Southern Appalachians

This notebook explores data from the following publication:

> **Stout, J.B., Jessee, L.D., & McMeen, J.N. (2025).** Nested and island models for determining the species-area relationship of southern Appalachian amphibians. *Journal of North American Herpetology*, 2025(1), 1-7. [https://doi.org/10.17161/jnah.v2025i1.21557](https://doi.org/10.17161/jnah.v2025i1.21557)

**Abstract:** Amphibians are widespread vertebrates in the temperate and tropical regions of the world that are facing global existential threats. The southern Appalachian region of eastern North America is an important stronghold for temperate amphibians, representing the global biodiversity hotspot for salamander genera and includes high rates of endemism. Despite a rich history of sampling in the region, a species-area relationship (SAR) for amphibians has not been attempted. We used nested and island site data from the southern Appalachian ecoregion to create linear regressions of two SAR models for predicting amphibian species richness. Our results indicate that land area is an accurate predictor of amphibian species richness within the study area. This model offers a rapid assessment of amphibian diversity and should be useful in conservation and land management initiatives. Identifying baseline biodiversity trends is vital for understanding species distributions into an uncertain future.

---

The SAR is a fundamental concept in ecology and biogeography that describes the relationship between the number of species present in a given area and the size of that area. The SAR is often depicted as a curve, known as the Species-Area Curve (SAC). The SAR states that as the size of a habitat or ecosystem increases, the number of species present in that area also increases, but at a decreasing rate. In other words, the more area you have, the more species you're likely to find. However, the rate of increase slows down as the area gets larger.

Here are some code examples that explore the data used in this project. First, let's import some libraries we will use in the code blocks below.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Now, load the data for this project into a Pandas DataFrame and display the frame.

```python
df = pd.read_csv('data/csv/species_area.csv')
display(df)
```

Now we can create a scatter plot to visualize the amphibian data.

```python
# plot the scatter plot using pyplot and the Pandas buildtin wrapper for pyplot
df.plot.scatter(x='Area', y='AmphibianSpecies', color='black')
plt.show()
```

To model the SAR, species richness (independent variable) and geographical area (dependent variable) samples were logarithmically (log10) transformed, and a regression analysis was performed to determine the statistical relationship between the variables. Regression modeling provided theoretical slope (z) and intercept (C) values for the formula $S= CA^z$ or in log-log form as $log(S) = log(C) + zlog(A)$ (Arrhenius 1921). Here is an example of the process.

```python
# create and populate log10 columns based on existing columns
df['Log10Area'] = np.log10(df['Area'])
df['Log10AmphibianSpecies'] = np.log10(df['AmphibianSpecies'])

# fit a linear regression line and getting the slope and y-intercept
m_amphibians, b_amphibians = np.polyfit(df['Log10Area'], df['Log10AmphibianSpecies'], 1)

# add the series values and the regression line to the scatter plot
fig, ax = plt.subplots()
df.plot.scatter(x='Log10Area', y='Log10AmphibianSpecies', color='black', ax=ax)
plt.plot(df['Log10Area'], m_amphibians * df['Log10Area'] + b_amphibians, color='black')
plt.show()

# Print the model derived from the regression
print(f'regression line: y = {m_amphibians:0.4f}x + {b_amphibians:0.4f}')
```

Calculating the regresssion allows us to fit the linear model $log(S) = log(C) + zlog(A)$. We arrive at the following model on logarithmic axes as $log(S) = 1.1437 + 0.1079 * log(A)$.

Let's create a predicitve function using these values, and plot the graph, which is just the regression line from the previous scatter plot. Then we will test the method for a value of $100km^2$

```python
# define our linear model
def amphibians(log_area):
    return 0.1079 * log_area + 1.1437

# create an array of 100 areas from 1.0 to 5.0
areas = np.linspace(1.0, 5.0, num = 100, endpoint=True)

# for each area calculate the log10 species
amphibian_species = [ amphibians(x) for x in areas ]

# plot the data
plt.plot(areas,amphibian_species, color='black')
plt.show()

# Let's make an estimate for 100km^2
print("Calculating estimated species for 100km^2")

# convert area to log10
log_area = np.log10(100)

# calculate log species
amphibian_species = amphibians(log_area)

# caculate species
amphibian_species = 10 ** amphibian_species

print(f'Species: {amphibian_species:0.1f}')
```

Given the previous model we can show $S= CA^z$ as $S =13.9219A ^ {0.1079}$. Let's plot this function on arithmetic axes.

```python
# define our model
def amphibians(area):
    return 13.9219 * area ** 0.1079

# create an array of 2000 areas from 1.0 to 2000.0
areas = np.linspace(1.0, 2000.0, num = 2000, endpoint=True)

# for each area calculate the species
amphibian_species = [ amphibians(x) for x in areas ]

# plot the data
plt.plot(areas,amphibian_species, color='black')
plt.show()

# Let's make an estimate for 100km^2
print("Calculating estimated species for 100km^2")

# calculate log species
amphibian_species = amphibians(100)

print(f'Species: {amphibian_species:0.1f}')
```

Our dataset breaks out amphibian species into frogs and salamanders, so we can calculate SAR models for each, and plot them alongside the linear curve for all Amphibians.

```python
# create and populate log10 columns based on existing columns
df['Log10FrogSpecies'] = np.log10(df['FrogSpecies'])
df['Log10SalamanderSpecies'] = np.log10(df['SalamanderSpecies'])

# fitting a linear regression line and getting the slope and y-intercept
m_frogs, b_frogs = np.polyfit(df['Log10Area'], df['Log10FrogSpecies'], 1)
m_salamanders, b_salamanders = np.polyfit(df['Log10Area'], df['Log10SalamanderSpecies'], 1)

# adding the regression line to the scatter plot
fig, ax = plt.subplots()
df.plot.scatter(x='Log10Area', y='Log10AmphibianSpecies', color='black', ax=ax)
df.plot.scatter(x='Log10Area', y='Log10FrogSpecies', color='green', ax=ax)
df.plot.scatter(x='Log10Area', y='Log10SalamanderSpecies', color='blue', ax=ax)
plt.plot(df['Log10Area'], m_amphibians * df['Log10Area'] + b_amphibians, color='black')
plt.plot(df['Log10Area'], m_frogs * df['Log10Area'] + b_frogs, color='green')
plt.plot(df['Log10Area'], m_salamanders * df['Log10Area'] + b_salamanders, color='blue')
plt.show()

# Print the model derived from the regression
print(f'amphibian regression line: y = {m_amphibians:0.4f}x + {b_amphibians:0.4f}')
print(f'frog regression line: y = {m_frogs:0.4f}x + {b_frogs:0.4f}')
print(f'salamander regression line: y = {m_salamanders:0.4f}x + {b_salamanders:0.4f}')
```

Given the regression lines we can caculate the C and z values for the power model $S= CA^z$

|Group|C|z|
|---|---|---|
|Amphibians|13.9219|0.1079|
|Frogs|4.2785|0.1338|
|Salamanders|8.9474|0.1043|

Let's plot the SAR for each group in arithmetic space.

```python
# define our models
def amphibians(area):
    return 13.9219 * area ** 0.1079

def frogs(area):
    return 4.2785 * area ** 0.1338

def salamanders(area):
    return 8.9474 * area ** 0.1043

# create an array of 2000 areas from 1.0 to 2000.0
areas = np.linspace(1.0, 2000.0, num = 2000, endpoint=True)

# for each area calculate the estimated species
amphibian_species = [ amphibians(i) for i in areas ]
frogs_species = [frogs(i) for i in areas ]
salamander_species = [salamanders(i) for i in areas ]

# plot the data
plt.plot(areas,amphibian_species)
plt.plot(areas,frogs_species)
plt.plot(areas,salamander_species)
plt.show()

# Let's make an estimate for 100km^2
print("Calculating estimated species for 100km^2")

# calculate species
amphibian_species = amphibians(100)
frog_species = frogs(100)
salamander_species = salamanders(100)

# print the results
print(f'Amphibians: {amphibian_species:0.1f}')
print(f'Frogs: {frog_species:0.1f}')
print(f'Salamanders: {salamander_species:0.1f}')
```

## Summary

This notebook demonstrates the species-area relationship (SAR) for amphibians in the southern Appalachian ecoregion, based on the findings from Stout, Jessee, & McMeen (2025). Key findings from the publication include:

- **Land area is an accurate predictor** of amphibian species richness within the southern Appalachian study area
- The study used both **nested** (increasingly larger concentric areas) and **island** (discrete habitat patches) sampling approaches to create linear regressions
- The power model $S = CA^z$ provides a reliable framework for predicting amphibian diversity:
  - **Amphibians:** $S = 13.92A^{0.1079}$
  - **Frogs:** $S = 4.28A^{0.1338}$
  - **Salamanders:** $S = 8.95A^{0.1043}$
